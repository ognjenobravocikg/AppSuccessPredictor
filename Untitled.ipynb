```python
import pandas as pd
import numpy as np
import sklearn as sk
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("Google-Playstore.csv")
```

```python
# Pretvaranje Boolova u intove
data['Ad Supported'] = data['Ad Supported'].astype(int)
data['In App Purchases'] = data['In App Purchases'].astype(int)
data['Editors Choice'] = data['Editors Choice'].astype(int)
data['Free'] = data['Free'].astype(int)
```

```python
data.head()
```

```python
data.drop(columns=['Rating Count', 'Released', 'Last Updated', 'Developer Email', 'Developer Website', 'Privacy Policy'], inplace=True)
```

```python
data.head()
```

```python
missingValues = data.isnull().sum()
missingValues
```

```python
noName = data[data['App Name'].isnull()]
noName
```

```python
data = data[data['App Name'].notna()]
data = data[data['Developer Id'].notna()]
```

```python
noInstalls = data[data['Installs'].isnull()]
noInstalls
```

```python
data.loc[data['Installs'].isnull(), 'Installs'] = data["Maximum Installs"] / 2
```

```python
numeric_columns = data.select_dtypes(include='number')
correlation = numeric_columns.corr()
correlation
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".4f",linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
